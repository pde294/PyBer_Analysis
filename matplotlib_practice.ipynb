```python
%matplotlib inline
```

```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
#Object oriented approach to 5.1.4
# Create the plot
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
# Create the plot. Change the line color to green. Change the marker to diamond. Change the line width to 2 points. 
plt.plot(x_axis, y_axis, marker=">", color="green", linewidth=2, label = 'Boston')
fig, ax = plt.subplots(figsize=(45, 45))

#Add a legend with the label "Boston"
#ax.legend()
ax.set_label ("Boston")
#Add a title and axes labels
ax.set_xlabel('Date')
ax.set_ylabel('Fares')
ax.set_title("PyBer Fare by Month")


#Add a y-axis limits

#Add grid lines
ax.grid()
ax.legend()
```

```python
fig, ax = plt.subplots()
# made alpha larger than 20% so I could see the bubbles better
ax.scatter(y_axis, x_axis, s=[i * 5 for i in y_axis], label = 'Boston', alpha =.5, edgecolors = "k")
ax.set_color = ("skyblue")
plt.title = ("PyBer Fare by Month")
plt.legend()
ax.set_ylabel('Fare')
ax.set_xlabel('Date')
plt.gca().invert_yaxis()
plt.xlim(0, 45)
#alpha(.2)
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```
